# Impute the missing age with the median age of the respective passenger class

```python
median_overall = titanic['Age'].median()
median_overall
```

```python
titanic_new = titanic.copy()
```

```python
titanic_new['Age'].fillna(median_overall,inplace=True)
```

```python
titanic_new
```

```python
titanic_new[titanic_new['Age'].isnull()]
```

```python
titanic.loc[5]
```

```python
titanic_new.loc[5]
```

```python
titanic_new1 = titanic.copy()
```

```python
# determine the median age of the passengers in various passenger class categories
med_age_class_1 = titanic[titanic['Pclass']==1]['Age'].median()
med_age_class_2 = titanic[titanic['Pclass']==2]['Age'].median()
med_age_class_3 = titanic[titanic['Pclass']==3]['Age'].median()
```

```python
med_age_class_1, med_age_class_2, med_age_class_3
```

```python
titanic_new1.loc[(titanic_new1.Age.isnull()) & (titanic['Pclass']==3),"Age"] = med_age_class_3
```

```python
titanic_new1.loc[5]
```

```python
titanic['Age_band']=0
titanic.loc[titanic['Age']<=16,'Age_band']=0
titanic.loc[(titanic['Age']>16)&(titanic['Age']<=32),'Age_band']=1
titanic.loc[(titanic['Age']>32)&(titanic['Age']<=48),'Age_band']=2
titanic.loc[(titanic['Age']>48)&(titanic['Age']<=64),'Age_band']=3
titanic.loc[titanic['Age']>64,'Age_band']=4
```
